**Hint**: достаточно просто взять group by по указанным выше категориальным колонкам и верно вставить полученный результат в исходную табличку.

```python
### Your code is here
cols_group = ['vendor_id', 'store_and_fwd_flag' , 'traffic_jam', 'month_2', 'month_3',
              'month_4', 'month_5', 'month_6']

mean_for_groups = data.groupby(cols_group)['distance_km'].transform('mean')
data['distance_km'].fillna(mean_for_groups, inplace=True)
```

### Теперь наши данные чисты! Замерьте качество модели на Кросс-Валидации

```python
from sklearn.model_selection import KFold

selector = KFold(n_splits=4, shuffle=True, random_state=33)

from sklearn.model_selection import cross_validate
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_log_error

### Your code is here
model = LinearRegression()

x = data.drop(columns='log_trip_duration')
y = data['log_trip_duration']

cross_val = cross_validate(model, x, y,
                           scoring='neg_mean_squared_error',
                          cv=selector, return_train_score=True)

cross_val_error = -cross_val['test_score'].mean()

print(f"MSLE на Кросс-валидации: {round(cross_val_error, 3)}")
```

Давайте почистим их, удалив те объекты, таргеты которых находятся вне промежутка

(топ 99,5% значение;   топ 0,5% значение)

```python
### Your code is here
lower_bound = data['log_trip_duration'].quantile(0.005)
upper_bound = data['log_trip_duration'].quantile(0.995)

cleaned_data = data[(data['log_trip_duration'] > lower_bound) & 
                    (data['log_trip_duration'] < upper_bound)].copy()

cleaned_data.head()
```

Снова замерьте качество на кросс-валидации, используя старый selector

```python
### Your code is here
model = LinearRegression()

x = cleaned_data.drop(columns='log_trip_duration')
y = cleaned_data['log_trip_duration']

cross_val = cross_validate(model, x, y,
                           scoring='neg_mean_squared_error',
                          cv=selector, return_train_score=True)

cross_val_error = -cross_val['test_score'].mean()

print(f"MSLE на Кросс-валидации: {round(cross_val_error, 3)}")
```
